```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = np.linspace(-1,1,50), np.linspace(-1,1,50)
X,Y = np.meshgrid(x,y)

C = 1
L = 1
n = 1


def d_x(x_,y_):
    d_x = C*(L/n)*(x_**3 + x_*y_**2)
    return d_x
def d_y(x,y):
    return C*(L/n)*(x**2*y+ y**3)

x_prime = []
y_prime = []

for i in x:
    x_prime.append(i+d_x(i,j))
for j in y:
    y_prime.append(i+d_y(i,j))

X_,Y_ = np.meshgrid(x_prime,y_prime)
```

```python
plt.figure(figsize=[20,20])
#plt.plot(x_prime, y_prime, 'k.')
plt.subplot(2,2,1)
plt.plot(X, 'r.', markersize = 10)
plt.plot(X, 'k-')
plt.subplot(2,2,2)
plt.plot(X_, 'r.', markersize = 10)
plt.plot(X_, 'k-')
plt.subplot(2,2,3)
plt.imshow(Y, cmap = 'seismic')
plt.subplot(2,2,4)
plt.imshow(Y_, cmap = 'seismic')
```
