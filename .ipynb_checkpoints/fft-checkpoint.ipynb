```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.fftpack as fft
from PIL import Image, ImageDraw, ImageFont
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
```

## Activity 1. Familiarization with discrete FFT

```python
t = np.linspace(-1, 1, 500)
X, Y = np.meshgrid(t,t)
R = np.hypot(X, Y)
Z = np.zeros((500,500))
Z[R<=0.5] = 1
mp.imshow(Z, "gray")
mp.show()
```

```python
FZ = fft.fft2(Z)
mp.imshow(abs(FZ).astype("uint8"), "gray")
mp.show()
```

```python
mp.imshow(fft.fftshift(abs(FZ)).astype("uint8"), "gray")
mp.show()
```

```python
mp.imshow(abs(fft.fft2(FZ)).astype("uint8"), "gray")
mp.show()
```
