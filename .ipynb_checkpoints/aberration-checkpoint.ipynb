```python
import numpy as np
import matplotlib.pyplot as mp
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["figure.figsize"] = (7*16/9,7)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["font.family"] = "serif"
```

```python
def len_make(S, R, n, np):
    Sprime = 1/((np - n)/R - n/S) * np
    return Sprime

def Cmake(S, Sprime, R, n, np):
    c1 = (n * (1/S + 1/R))**2/n/S
    c2 = (np * (1/Sprime - 1/R))**2/np/Sprime
    return c1 + c2
    
def I(rho, S, R, n, np):
    Sprime = len_make(S, R, n, np)
    C = Cmake(S, Sprime, R, n, np)
    return -C*rho**4/8
```

```python
rho = np.linspace(-1e6,1e6,500)
fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(rho, I(rho=rho, S=10, R=35e-3, n=1.33, np=1.00), "b-")
ax.set_xlabel(r"$\rho$")
ax.set_ylabel(r"$I(\rho)$")
mp.savefig("I_rho.png", bbox_inches="tight", dpi=300)
mp.show()
```
