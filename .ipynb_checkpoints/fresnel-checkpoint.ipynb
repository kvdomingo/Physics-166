```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.fftpack as fft
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5, 5)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["font.family"] = "serif"
```

```python
res = 256
for rad in range(5, res//2, 50):
    #rad = 50
    t = np.arange(-res//2, res//2, 1)
    X, Y = np.meshgrid(t,t)
    R = np.hypot(X, Y)
    Z = np.ones((res, res))
    Z[R <= rad] = 0
    Z = np.round(Z/Z.max() * 255).astype("uint8")

    M = Z * np.exp(1j*((X**2 + Y**2)*(np.pi/res)))
    I = abs(fft.fftshift(fft.fft2(M)))

    fig = mp.figure(figsize=(5*3,5))
    
    ax = fig.add_subplot(131)
    ax.imshow(Z, "gray")
    ax.set_title("aperture radius = %i"%rad)

    ax = fig.add_subplot(132)
    ax.imshow(I**2, "gray")
    ax.set_title("Fresnel diffraction pattern")

    ax = fig.add_subplot(133)
    mp.plot(I[res//2], "k-")
    ax.set_title("line scan")
    
    mp.show()
```
