```python
import numpy as np
import matplotlib.pyplot as mp
from mpl_toolkits.mplot3d import Axes3D
from jupyterthemes import jtplot
```

```python
jtplot.style()
mp.style.use("seaborn")
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
```

```python
fig = mp.figure()

ax = fig.add_subplot(111, projection="3d")
ax.view_init(azim=30, elev=45)
t = np.linspace(-2*np.pi, 2*np.pi, 1000)
X,Y = np.meshgrid(t,t)
R = np.hypot(X,Y)
Z = np.sinc(R)
ax.plot_surface(X, Y, Z, cmap="hot")

ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.set_zlim(-1, 1)

ax.contour(X, Y, Z, zdir="x", offset=-8, cmap="hot")
ax.contour(X, Y, Z, zdir="y", offset=-8, cmap="hot")
ax.contour(X, Y, Z, zdir="z", offset=-1, cmap="hot")

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")

mp.show()
```
